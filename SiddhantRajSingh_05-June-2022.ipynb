####  (i)	25% on the final price (discounted price) of total orders if the number of orders placed is greater than 600

####  (iii) 10% on the final price (discounted price) of total orders if the number of orders placed is greater than 100 but less than or equal to 350 
####  (iv) Otherwise, 5% on the final price (discounted price) of total orders

## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5

#### Creating a method called 'user_method' which will take every value of the column i.e. the discounted price of the order and pass it as a parameter to calculate the ECommerce revenue after applying the rules by which the company charges each retailer

````python
def user_method(x):
    if x > 600:
        return x * 0.25
    elif x > 350 and x <= 600:
        return x * 0.15
    elif x > 100 and x <= 350:
        return x * 0.10
    elif x <=100:
        return x * 0.05
````

#### using apply() method on the column 'discounted_price' to get the ECommerce revenue and creating a new column 'ECom_revenue' to store the ECommerce revenue of each order 

````python
ECom_Data['ECom_revenue'] = ECom_Data['discounted_price'].apply(user_method)
````

#### the net revenue generated by the E-Commerce company over all orders placed can be calculated using the code
```python
ECom_Data['ECom_revenue'].sum()
```

````python
ECom_Data['ECom_revenue'].sum()
````

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of ECom Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)


### Ans 9 a)

#### Converting the column 'Delivered_Date' into 'datetime' datatype

````python
ECom_Data['Order_Date']= pd.to_datetime(ECom_Data['Delivered_Date'],format='%d-%m-%Y',errors ='coerce')
````

#### Checking the Datatype of the column 'Delivered_Date'

````python
ECom_Data['Order_Date'].dtype
````

#### Extracting the Month name from the column 'Delivered_Date'

````python
ECom_Data['month_name'] = ECom_Data['Order_Date'].dt.month_name()
````

#### Checking the month_name column in DataFrame

````python
ECom_Data
````

#### applying groupby to columns 'Region' and 'month_name' in order to get the monthly Revenue of ECom Company for each region separately.

````python
monthly_data_per_region = ECom_Data.groupby(by = ['Region','month_name']).sum()['ECom_revenue']
````

#### Converting Series into DataFrame

````python
monthly_data_per_region = pd.DataFrame(monthly_data_per_region)
````

````python
monthly_data_per_region
````

````python
plt.figure(figsize = (12,9), dpi = 200)
sns.lineplot(data = monthly_data_per_region, x='month_name', y='ECom_revenue', hue = 'Region',ci=False, palette='Dark2')
plt.xticks(rotation = 90, fontsize = 8)
plt.xlabel('Month Name',labelpad = 8, fontsize = 12)
plt.ylabel('ECommerce Revenue', labelpad = 8, fontsize = 12)
plt.show()
````
